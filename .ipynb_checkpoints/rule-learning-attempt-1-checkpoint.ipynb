```python
# first neural network with keras tutorial
import numpy as np
import keras
import math
from keras.models import Sequential
from keras.layers import Dense
from pandas import read_csv
from matplotlib import pyplot
```

```python
# load the dataset
#dataset = loadtxt('pima-indians-diabetes.csv', delimiter=',')
#dataset = read_csv('move-right-basic.csv')
#dataset = read_csv('hardcoded-labels-pythagorean.csv')
dataset = read_csv('data/addition.csv')
```

```python
features = dataset.iloc[:, 0:1].values
labels = dataset.iloc[:, 1:2].values
```

```python
# split data into train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.98, random_state=0)
```

```python
import keras.backend as K


def euclidean_distance_loss(y_true, y_pred):
    """
    Euclidean distance loss
    https://en.wikipedia.org/wiki/Euclidean_distance
    :param y_true: TensorFlow/Theano tensor
    :param y_pred: TensorFlow/Theano tensor of the same shape as y_true
    :return: float
    """
    return K.sqrt(K.sum(K.square(y_pred - y_true), axis=-1))
```

```python
# define the keras model
from keras import layers
from keras.layers import LeakyReLU
from keras.layers import Lambda
from keras.initializers import Constant

# -------- LEARNING FUNCTIONAL API

from keras.layers import Input, Dense
from keras.models import Model

# This returns a tensor
inputs = Input(shape=(1,))

# a layer instance is callable on a tensor, and returns a tensor
firstLeftOut = LeakyReLU(alpha=0.1)(Dense(1)(inputs))
firstRightOut = LeakyReLU(alpha=0.1)(Dense(1)(inputs))
secondInput = keras.layers.concatenate([firstLeftOut, firstRightOut])
predictions = Dense(1, activation='relu')(secondInput)

# This creates a model that includes
# the Input layer and three Dense layers
model = Model(inputs=inputs, outputs=predictions)

# -------- ORIGINAL SEQUENTIAL MODEL

#model.add(layers.Dense(1, input_dim=1, kernel_initializer=Constant(value=0)))

#model = Sequential()
#model.add(layers.Dense(1, input_dim=1))
#model.add(LeakyReLU(alpha=0.1))

#model.add(Lambda(lambda x: x + 1))

#model.add(layers.Dense(8, input_dim=2, activation='relu'))
#model.add(layers.Dense(12, activation='relu'))
#model.add(layers.Dropout(rate=0.25, input_shape=(2,)))
#model.add(layers.Dense(10, activation='relu'))
#model.add(layers.Dropout(rate=0.5, input_shape=(2,)))
#model.add(layers.Dense(3, activation='relu'))
#model.add(Dense(8, activation='relu'))

#model.add(layers.Dense(1))
#model.add(LeakyReLU(alpha=0.1))
```

```python
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot

SVG(model_to_dot(model).create(prog='dot', format='svg'))
```

```python
model.summary()
```

```python
for layer in model.layers:
    if len(layer.get_weights()) > 0:
        print("Weights:")
        print(layer.get_weights()[0])
        print("Biases:")
        print(layer.get_weights()[1])
```

```python
# compile the keras model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
tbCallback = keras.callbacks.TensorBoard(log_dir='./logs', histogram_freq=1, batch_size=400, write_graph=True, write_grads=True, write_images=False, embeddings_freq=0, embeddings_layer_names=None, embeddings_metadata=None, embeddings_data=None, update_freq='epoch')
```

```python
# fit the keras model on the dataset
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=5000, batch_size=400, shuffle=True)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X_train, y_train)
print('Accuracy on Train: %.2f' % (accuracy*100))
_, accuracy = model.evaluate(X_test, y_test)
print('Accuracy on Test: %.2f' % (accuracy*100))
```

```python
# Plot training & validation accuracy values
pyplot.plot(history.history['acc'])
pyplot.plot(history.history['val_acc'])
pyplot.title('Model accuracy')
pyplot.ylabel('Accuracy')
pyplot.xlabel('Epoch')
pyplot.legend(['Train', 'Test'], loc='upper left')
pyplot.show()

# Plot training & validation loss values
pyplot.plot(history.history['loss'])
pyplot.plot(history.history['val_loss'])
pyplot.title('Model loss')
pyplot.ylabel('Loss')
pyplot.xlabel('Epoch')
pyplot.legend(['Train', 'Test'], loc='upper left')
pyplot.show()
```
